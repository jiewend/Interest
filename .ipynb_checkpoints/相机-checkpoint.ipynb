## 相机参数，[参考知乎](https://www.zhihu.com/question/24784518)
- ### 景深：清晰成像的距离
    - 景深小，前景清晰，后景模糊，可用于人像中模糊背景
    - 景深大，前景清晰，后景也清晰，
- ### 光圈：控制镜头进光量，常见为“F数字”
    - 数字越小，光圈越大，进光越多，景深越小，背景模糊
    - 相机模式“光圈优先（A）”：光圈用于控制景深，大光圈可以失焦背景，强调人物
- ### 快门速度：快门打开到关闭所用时间
    - 快门速度高，曝光时间越短，进光量越少，画面越暗
    - 相机模式“快门优先（S）”：手动设定快门速度，光圈大小由相机自动配合
    - 常见快门设置
        - 人物重影效果：1秒，1/2秒等，拍摄刀光剑影，相对模糊且有意境的照片
        - 手持拍摄：不低于1/60秒，不然可能会出现相机不稳，画面模糊
        - 行走或快速移动：1/250秒，可以得到相对清晰，并没有抖动的画面
- ### ISO感光度：调节可捕捉光量
    - 快门速度不变时，提高ISO可以增大光圈
    - ISO越高，拍出的照片噪点越多
    - 普通单反ISO800以下画质不错，白天拍风景控制在100-200
- ### 焦距：决定拍摄成像的属性
    - 焦距长短决定拍摄成像大小，视场角大小，景深大小和画面透视强弱
    - 焦距越长，你和被拍物体之间的距离越近，景深越浅，背景越模糊
    - 镜头上数字，18-55mm，表示此镜头焦距为18-55mm，f3.6-5.6，表示光圈，焦距最小18mm时，光圈最大f3.5；焦距最大55mm时，最大光圈f5.6
- ### 曝光度EV：反映曝光多少的量
    - 曝光过度，画面会发白，曝光过度，画面变暗
    - 很多单反都有自动曝光功能，但是在一些特殊情况下（逆光），需要人工增减EV
